<h3>Задание 3</h3>

Импортируем библиотеки nltk, pymorphy2, json. Открываем наш текстовый файл в режиме чтения. Сначала убираем пунктуацию и токенизируем текст. Потом создаем словарь и список, в который мы потом будем класть словари с леммами, словами, частями речи. В конце цикла получаем список, который потом сохраняем в json.

```python
#импортируем библиотеки nltk, pymorphy2, json
from nltk.tokenize import word_tokenize
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
import json
import re
#открываем текст
with open('Понятие и трагедия культуры.txt') as f:
    full_book = f.read()
#токенизируем текст
full_book = re.sub(r'[.,"\'-?:!;{}]', '', full_book)
tokens = word_tokenize(full_book)
#создаем словарь и список, в который потом будем класть наши словари
main_dict={}
list_dict = []
for word in tokens:
    main_dict['lemma'] = morph.parse(word)[0].normal_form
    main_dict['word'] = morph.parse(word)[0].word
    main_dict['POS'] = morph.parse(word)[0].tag.POS
    list_dict.append(main_dict)
    main_dict = {}

print(list_dict)
#записываем результат в файл json
with open("data.json", 'w') as f:
    for item in list_dict:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")
```

<h3>Задание 4</h3>

Мы создаем пустой список и словарь. В список мы будем класть наши части речи, которые встречаются в тексте. Потом через словарь посчитаем частоту каждой части речи. Это мы сейчас сделали первый пункт задачи четыре, где нужно было найти долю слов, которую составляет каждая часть речи.

```python
#импортируем нужные библиотеки 
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()


with open('Понятие и трагедия культуры.txt') as f:
    text = f.read()

parts_in_text = [] #создаем пустой список
freq = {} #создаем пустой словарь
#добавляем части речи какие встречаются в тексте в наш список
for word in text:
    parts_in_text.append(morph.parse(word)[0].tag.POS)

for part in parts_in_text:
    if part in freq: #если слово уже входит в словарь, то увеличим его частоту
        freq[part] += 1
    else:           #если не входит, то создадим запись в словарь
        freq[part] = 1

print(freq) #распечатаем словарь с частотой встречаемости чати речи
print(len(text)) #для удобства распечатаем длину нашего текста
for k,v in freq.items():   # для каждой части речи теперь посчитаем ее долю
    dolya = v/len(text)
    print('Средняя доля' ,k , ':' , dolya)
```

<h3>Задание 4</h3>

Это продолжение задания 4, второй пункт, где нужно найти топ-20 (по частотности) глаголов и наречий. Сначала лемматизируем текст, чтобы была начальная форма каждого слова, то есть вместо глаголов будем смотреть их инфинитивы. Создаем пустые списки, в которые потом будем добавлять наречия и инфинитивы. Через словари находим частоту встречаемости глаголов и наречий. Сортируем ключи и распечатываем результат.

```python
#импортируем нужные библиотеки
from pymystem3 import Mystem
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
m = Mystem()
#открываем текстовый файл в режиме чтения
with open('Понятие и трагедия культуры.txt') as f:
    full_book = f.read()
#лемматизируем текст
text = m.lemmatize(full_book)
#создаем для наречий и глаголов пустые списки
list_verbs = []
list_advb = []
for word in text:
    if 'INFN' in morph.parse(word)[0].tag:
        list_verbs.append(word)
    if 'ADVB' in morph.parse(word)[0].tag:
        list_advb.append(word)
#создаем пустые словари
freq_verbs = {}
freq_advb = {}
#через словарь считаем частоту глаголов и наречий
for vb in list_verbs:
    if vb in freq_verbs:
        freq_verbs[vb] += 1
    else:
        freq_verbs[vb] = 1

for av in list_advb:
    if av in freq_advb:
        freq_advb[av] += 1
    else:
        freq_advb[av] = 1
#отсортируем словари по ключам по убыванию
sorted_verbs = sorted(freq_verbs.items(), key=lambda kv: (-kv[1], kv[0]))
sorted_advb = sorted(freq_advb.items(), key=lambda kv: (-kv[1], kv[0]))
#распечатаем каждый
print(sorted_verbs[0:20])
print(sorted_advb[0:20])
```

И так проверям с каждым глаголом в прошедшем, настоящем и будущем времени. После этого меняем время у глаголов, чтобы согласовать их по числу с существительными. Выводим получивший текст. 

```python
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
from string import punctuation


abzac = """На основании первого глубинного дуализма, - а именно того факта, что человек не включается безусловно,
 подобно зверю, в естественную мировую заданность, но от нее отделяется, противопоставляет себя ей, требуя и борясь,
  одолевая ее и будучи одолеваем ею, - происходит завязывание бесконечного взаимодействия между субъектом и объектом.
   Следующее по порядку проявление этого процесса имеет место внутри самого духа. Дух производит бесчисленные порождения,
    которые продолжают свое собственное, самостоятельное существование, независимое как от души, их сотворившей, так и от
     любой другой воспринимающей или же отклоняющей их души. Субъект оказывается в таком положении как в отношении искусства,
      так и права, как религии, так и техники, как науки, так и морали: их содержание не просто то притягивает его, а то
       отталкивает, то оказывается сплавленным с ним воедино, как некая часть его собственного Я, а то противостоящим ему
        в отчужденности и отсутствии каких-либо точек соприкосновения. Нет, скорее можно сказать, что здесь имеет место 
        некая косность, окаменелость, пребывающая экзистенция, с которой дух, превратившись таким образом в объект,
         противостоит текучей живости, внутренней ответственности, идущим друг другу на смену напряженностям души.
          Будучи, как дух, теснейшим образом связан с иным таким же духом, именно в силу этого он, однако, переживает
           бесчисленные трагедии, сопряженные с глубоким противоречием между той и другой формой: между субъективной
            жизнью, которая не знает покоя, однако конечна во времени, и ее содержанием, которое, будучи раз роздано,
             неподвижно, однако вечно сохраняет свое значение. В середине этого дуализма пребывает идея культуры.
              В основе ее лежит тот внутренний факт, который возможно выразить в полном виде только символически и в расплывчатой
               форме: как путь души к себе самой. Ибо душа никогда не является только лишь тем, что она есть в данный момент,
                но всегда есть нечто большее: в ней уже заранее задана ее собственная, более высокая и совершенная в сравнении
                 с нею же самой форма, и пусть в нереальном виде, как-то все же она тут присутствует. Речь идет здесь не об определенном,
                  зафиксированном в некоторой точке духовного мира идеале, но об освобождении покоящихся в самой душе напряженных сил,
                   развитии ее собственного, повинующегося ее же формообразующему побуждению зерна"""

text = [word.lower().strip(punctuation) for word in abzac.split()]
text = [word for word in text if word != '']
list = [] #создаем пустой список
#у каждого существительного изменим число и у каждого глагола поменяем время
for word in text:
    if "NOUN" in morph.parse(word)[0].tag:
        if "sing" in morph.parse(word)[0].tag:
            list.append(morph.parse(word)[0].inflect({'plur'}).word)
        elif "plur" in morph.parse(word)[0].tag:
            list.append(morph.parse(word)[0].inflect({'sing'}).word)
    elif "VERB" in morph.parse(word)[0].tag:
        if "past" in morph.parse(word)[0].tag:
            if morph.parse(word)[0].inflect({'pres'}) != 'None':
                list.append(morph.parse(word)[0].inflect({'pres'}).word)
            else:
                list.append(morph.parse(word)[0].inflect({'futr'}).word)
        elif "pres" in morph.parse(word)[0].tag:
            if morph.parse(word)[0].inflect({'past'}) != 'None':
                list.append(morph.parse(word)[0].inflect({'past'}).word)
            else:
                list.append(morph.parse(word)[0].inflect({'futr'}).word)
        elif "futr" in morph.parse(word)[0].tag:
            if morph.parse(word)[0].inflect({'pres'}) != 'None':
                list.append(morph.parse(word)[0].inflect({'pres'}).word)
            else:
                list.append(morph.parse(word)[0].inflect({'past'}).word)
    else:
        list.append(word)

spisok = []#создадим пустой список
#поменяем число у глаголов, чтобы они были согласованы с существительными по числу
for item in list:
    if "VERB" in morph.parse(item)[0].tag:
        if "sing" in morph.parse(item)[0].tag:
            spisok.append(morph.parse(item)[0].inflect({'plur'}).word)
        else:
            spisok.append(morph.parse(item)[0].inflect({'sing'}).word)
    else:
        spisok.append(item)

print(' '.join(spisok))
```
